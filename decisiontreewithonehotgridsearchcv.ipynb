```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
```

```python
df=pd.read_csv('/content/loan_default_risk_dataset.csv')
```

```python
df.dropna()
df.info()
```

```python
X = df.drop('Loan_Default_Risk',axis=1)
y  =df['Loan_Default_Risk']
```

```python

label = LabelEncoder()
df['Loan_Default_Risk'] = label.fit_transform(df['Loan_Default_Risk'])
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,y,train_size=0.8,random_state=42)
cat_col  = X.select_dtypes(include='object').columns
num_col  = X.select_dtypes(include=['int64','float64']).columns
```

```python
preprocessing = ColumnTransformer(
    transformers=[
       ('onehotencoder',OneHotEncoder(sparse_output=False),cat_col)
    ],remainder='passthrough'
)
```

```python
pipeline = Pipeline(
    steps=[
      ('preprocessing',preprocessing),
      ('decisiontree',DecisionTreeClassifier(random_state=42,criterion='entropy'))
    ]
)
```

```python
pipeline.fit(xtrain,ytrain)
```

```python
pipeline.score(xtest,ytest)
```

```python
params = {
    'Criterion' :['entropy','gini'],
    'max_depth' :[5,10,50,100,200],
    'min_samples_leaf' :[2,3,5,7,10],
    'min_samples_split' :[2,3,5,10]
}
```

```python
gridsearch = GridSearchCV (pipeline,params,cv=5,n_jobs=-1)
```
